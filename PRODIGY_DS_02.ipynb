```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(df2['Age'],bins=20,kde=True)
plt.title('Histogram of Numerical Column')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='',y='Age',data=df2)
plt.title('Box Plot of Numerical Column Grouped by Category Column')
plt.show()
```

```python
sns.pairplot(df2[['Age','SibSp','Parch']])
plt.title('Pair Plot of Selected Numerical Columns')
plt.show()
```
