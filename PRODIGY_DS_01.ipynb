```python
import pandas as pd
df2=pd.read_csv("D:\d3.csv")
df2
```

```python
import matplotlib.pyplot as plt
gender=df2['Region'].value_counts()
width=0.9
x=range(len(gender.index))
plt.bar(gender.index,gender.values)
plt.xlabel('region')
plt.ylabel('count')
plt.title("distribution of gender")
plt.xticks(x,gender.index,rotation=45)
plt.tight_layout()
plt.show
```
