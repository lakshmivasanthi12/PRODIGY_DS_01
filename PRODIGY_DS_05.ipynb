```python
import pandas as pd
file_path=r"D:\teslaa.csv"
df2=pd.read_csv(file_path)
print(df2)
```

```python
df2=df2.dropna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.countplot(x='Volume',data=df2)
plt.title('tesla data on stock prices')
plt.xlabel('Open')
plt.ylabel('Close')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Adj Close', data=df2)
plt.title('Rate of stock prices')
plt.xlabel('High')
plt.ylabel('Low')
plt.xticks(rotation=45)
plt.show()
```
